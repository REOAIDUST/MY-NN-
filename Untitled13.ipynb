```python
import numpy as np
```

```python
def f(x, y):
    return x**2 + y**2 - 4*x - 6*y

N = 3
max_iter = 3
w = 0.5
c1 = 1.5
c2 = 2


np.random.seed(0)
positions = np.random.uniform(-5, 5, (N, 2))
velocities = np.random.uniform(-1, 1, (N, 2))

pbest_pos = positions.copy()
pbest_val = np.array([f(x, y) for x, y in positions])

gbest_index = np.argmin(pbest_val)
gbest_pos = pbest_pos[gbest_index]

for it in range(max_iter):
    for i in range(N):
        r1, r2 = np.random.rand(), np.random.rand()

        velocities[i] = (
            w * velocities[i]
            + c1 * r1 * (pbest_pos[i] - positions[i])
            + c2 * r2 * (gbest_pos - positions[i])
        )

        positions[i] += velocities[i]

        value = f(positions[i][0], positions[i][1])

        if value < pbest_val[i]:
            pbest_val[i] = value
            pbest_pos[i] = positions[i]

    gbest_index = np.argmin(pbest_val)
    gbest_pos = pbest_pos[gbest_index]

# Result
print("Best position (x, y):", gbest_pos)
print("Minimum value:", f(gbest_pos[0], gbest_pos[1]))
```
